```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix

# Load the dataset
file_path = 'E:\Assignment\AI&ML\GYM.csv'
data = pd.read_csv(file_path)
```

```python
#Basic Information about the dataset
print("Basic Information:")
print(data.info())  # Overview of dataset

print("\nSummary Statistics:")
print(data.describe())  # Summary statistics
```

```python
#Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

#Distribution of numerical features
numerical_columns = ['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM', 'Resting_BPM', 
                     'Session_Duration (hours)', 'Calories_Burned', 'Fat_Percentage', 'Water_Intake (liters)', 
                     'Workout_Frequency (days/week)', 'BMI']

print("\nDistribution of Numerical Features:")
data[numerical_columns].hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
#Correlation Heatmap for numerical variables
print("\nCorrelation Heatmap:")
plt.figure(figsize=(12, 8))
corr_matrix = data[numerical_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# 7. Boxplot of key variables to check for outliers
print("\nBoxplots to Check for Outliers:")
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[numerical_columns], orient="h")
plt.title('Boxplot of Numerical Features to Check for Outliers')
plt.show()
```

```python
# 8. Workout Type vs. Calories Burned
print("\nCalories Burned by Workout Type:")
plt.figure(figsize=(8, 6))
sns.boxplot(x='Workout_Type', y='Calories_Burned', data=data)
plt.title('Calories Burned for Different Workout Types')
plt.show()
```
